```python
import pandas as pd
```

```python
scale =pd.read_csv('describe_all_scale.csv')
stat = pd.read_csv('describe_all_stat.csv')
```

```python
res = pd.concat([scale, stat], axis=0)
res.columns
```

```python

sort = ["sus", "effect", "easy", "easy_learn", "satisfaction", "mouse_distance", "mouse_click", "keyboard_press", "duration"]
res = res.sort_values(by=["column_name", "task_round"], key=lambda x: x.map({k: i for i, k in enumerate(sort)})).reindex(["column_name", "task_round", "sample_size", "mean", "std", "std_error", "ci_lower", "ci_upper", "min", "max"], axis=1).reset_index(drop=True)
res
```

```python
res["task_round"] = res["task_round"].map({ 0: "WIMP", 1: "M-CAT"})
```

```python
res = res.rename(columns={"column_name": "因变量", "task_round": "任务组（自变量）", "sample_size": "样本量", "mean": "均值", "std": "标准差", "std_error": "标准误", "ci_lower": "95% 置信区间下限", "ci_upper": "95% 置信区间上限", "min": "最小值", "max": "最大值"})
```

```python
res["因变量"] = res["因变量"].map({"sus": "SUS 评分", "effect": "有效性", "easy": "易用性", "easy_learn": "易学性评分", "satisfaction": "满意度", "mouse_distance": "鼠标移动距离", "mouse_click": "鼠标点击次数", "keyboard_press": "键盘按键次数", "duration": "任务时长", "cost": "操作成本"})
res  = res.round(2)
res
```

```python
res.to_csv("describe_all.csv", index=False, float_format="%.2f")
```
